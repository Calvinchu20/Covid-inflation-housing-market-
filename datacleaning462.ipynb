```python
import pandas as pd
```

```python
df = pd.read_csv("housingvalue.csv")
df = df.fillna(0)
df
```

```python
df =df.iloc[1:]
date_columns = df.columns[5:]
df[date_columns] = df[date_columns].apply(pd.to_numeric, errors='coerce')

# Function to extract year from column name
def extract_year(column_name):
    return int(column_name.split('-')[0])

# Create a dictionary to hold aggregated values for each year
yearly_data = {}

# Iterate through each column
for column in date_columns:
    year = extract_year(column)
    if year not in yearly_data:
        yearly_data[year] = []
    yearly_data[year].append(column)

# Create new columns for each year and populate them with aggregated values
for year, columns in yearly_data.items():
    df[str(year)] = df[columns].mean(axis=1)

# Drop the original date columns
df = df.drop(columns=date_columns)
df
```

```python
df = df[['RegionID',"SizeRank","RegionName","RegionType","StateName","2021","2022","2023","2024"]]
df
```

```python
df_melted = pd.melt(df, id_vars=['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName'], 
                    var_name='Year', value_name='HousingValue')

# Convert the 'Year' column to numeric
df_melted['Year'] = pd.to_numeric(df_melted['Year'])

# Sort the DataFrame by 'RegionName' and 'Year' for clarity
df_melted = df_melted.sort_values(by=['RegionName', 'Year'])

# Display the reshaped DataFrame
print(df_melted.head())
df_melted
```

```python
df_melted.to_csv("housingvalue(cleaned).csv")
```

```python
df1 = pd.read_csv("1bedvalue(un).csv")
df2 = pd.read_csv("2bedvalue(un).csv")
df3 = pd.read_csv("3bedvalue(un).csv")
df4 = pd.read_csv("4bedvalue(un).csv")
df5 = pd.read_csv("5bedvalue(un).csv")
```

```python
def extract_year(column_name):
    return int(column_name.split('-')[0])
def filterYear(df):
    date_columns = df.columns[5:]
    df[date_columns] = df[date_columns].apply(pd.to_numeric, errors='coerce')

    # Function to extract year from column name
    

    # Create a dictionary to hold aggregated values for each year
    yearly_data = {}

    # Iterate through each column
    for column in date_columns:
        year = extract_year(column)
        if year not in yearly_data:
            yearly_data[year] = []
            yearly_data[year].append(column)

    # Create new columns for each year and populate them with aggregated values
    for year, columns in yearly_data.items():
        df[str(year)] = df[columns].mean(axis=1)

    # Drop the original date columns
    df = df.drop(columns=date_columns)
    
    df = df[['RegionID',"SizeRank","RegionName","RegionType","2021","2022","2023","2024"]]
    return df
    
```

```python
df1 = filterYear(df1)
df2 = filterYear(df2)
df3 = filterYear(df3)
df4 = filterYear(df4)
df5 = filterYear(df5)
```

```python
def combineYear(df):
    df_melted = pd.melt(df, id_vars=['RegionID', 'SizeRank', 'RegionName', 'RegionType'], 
                    var_name='Year', value_name='HousingValue')

    # Convert the 'Year' column to numeric
    df_melted['Year'] = pd.to_numeric(df_melted['Year'])

    # Sort the DataFrame by 'RegionName' and 'Year' for clarity
    df_melted = df_melted.sort_values(by=['RegionName', 'Year'])

    # Display the reshaped DataFrame
    print(df_melted.head())
    return df_melted
```

```python
df1 = combineYear(df1)
df2 = combineYear(df2)
df3 = combineYear(df3)
df4 = combineYear(df4)
df5 = combineYear(df5)
```

```python
df1.to_csv("1bedroomvalue.csv")
df2.to_csv("2bedroomvalue.csv")
df3.to_csv("3bedroomvalue.csv")
df4.to_csv("4bedroomvalue.csv")
df5.to_csv("5bedroomvalue.csv")
```

```python
import pandas as pd
```

```python
def cleanData(df):
    df = df.fillna(0)
    df = df.iloc[1:]
    return df
```

```python
def extract_year(column_name):
    return int(column_name.split('-')[0])
def filterYear(df):
    date_columns = df.columns[5:]
    df[date_columns] = df[date_columns].apply(pd.to_numeric, errors='coerce')

    # Function to extract year from column name
    

    # Create a dictionary to hold aggregated values for each year
    yearly_data = {}

    # Iterate through each column
    for column in date_columns:
        year = extract_year(column)
        if year not in yearly_data:
            yearly_data[year] = []
            yearly_data[year].append(column)

    # Create new columns for each year and populate them with aggregated values
    for year, columns in yearly_data.items():
        df[str(year)] = df[columns].mean(axis=1)

    # Drop the original date columns
    df = df.drop(columns=date_columns)
    
    df = df[['RegionID',"SizeRank","RegionName","RegionType","StateName","2021","2022","2023","2024"]]
    return df
    
```

```python
def combineYear(df):
    df_melted = pd.melt(df, id_vars=['RegionID', 'SizeRank', 'RegionName', 'RegionType',"StateName"], 
                    var_name='Year', value_name='rename')

    # Convert the 'Year' column to numeric
    df_melted['Year'] = pd.to_numeric(df_melted['Year'])

    # Sort the DataFrame by 'RegionName' and 'Year' for clarity
    df_melted = df_melted.sort_values(by=['RegionName', 'Year'])

    # Display the reshaped DataFrame
    print(df_melted.head())
    return df_melted
```

```python
#percentabove 
df = pd.read_csv("percenthomesoldabove.csv")
df_above = cleanData(df)
df_above = filterYear(df_above)
df_above = combineYear(df_above)
df_above = df_above.rename(columns={"rename":"percentAbove"})
#percentbelow
df=pd.read_csv("percenthomesoldbelow.csv")
df_below = cleanData(df)
df_below = filterYear(df_below)
df_below = combineYear(df_below)
df_below = df_below.rename(columns ={"rename":"percentBelow"})
```

```python
df_percent = pd.merge(df_above,df_below, on =['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'Year'],how = "outer")
df_percent=df_percent.drop_duplicates()
df_percent
```

```python
#mean sale price 
df = pd.read_csv("meansalepricemonthly.csv")
df_meanSale = cleanData(df)
df_meanSale = filterYear(df_meanSale)
df_meanSale = combineYear(df_meanSale)
df_meanSale = df_meanSale.rename(columns={"rename":"meanSaleprice"})
df_meanSale
```

```python
df_pm = pd.merge(df_percent,df_meanSale,on =['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'Year'],how = "outer" )
df_pm
```

```python
# sale count
df = pd.read_csv("salecountnowcast.csv")
df_salecount = cleanData(df)
df_salecount = filterYear(df_salecount)
df_salecount = combineYear(df_salecount)
df_salecount = df_salecount.rename(columns ={"rename":"saleCountnowcast"})
# saletolistratio
df = pd.read_csv("saletolistratiomonthlymean.csv")
df_saletolist = cleanData(df)
df_saletolist = filterYear(df_saletolist)
df_saletolist = combineYear(df_saletolist)
df_saletolist = df_saletolist.rename(columns = {"rename":"saletolist"})
df_sale = pd.merge(df_saletolist,df_salecount,on =['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'Year'],how = "inner")
df_sale
```

```python
combine_df = pd.merge(df_pm,df_sale,on =['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'Year'])
combine_df
```

```python
#forsale
df = pd.read_csv("forsaleinventorymontly.csv")
df_invent = cleanData(df)
df_invent = filterYear(df_invent)
df_invent = combineYear(df_invent)
df_invent = df_invent.rename(columns={"rename":"inventory"})
df = pd.read_csv("medianlistpricemontly.csv")
df_medianlist = cleanData(df)
df_medianlist = filterYear(df_medianlist)
df_medianlist = combineYear(df_medianlist)
df_medianlist = df_medianlist.rename(columns = {"rename":"medianlistprice"})
df = pd.read_csv("newlistingmontly.csv")
df_newlist = cleanData(df)
df_newlist = filterYear(df_newlist)
df_newlist = combineYear(df_newlist)
df_newlist = df_newlist.rename(columns={"rename":"newlist"})

df_for = pd.merge(df_invent,df_medianlist,on =['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'Year'])
df_forSale = pd.merge(df_for,df_newlist,on =['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'Year'])
df_price = pd.merge(df_forSale,combine_df,on =['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'Year'])
df_price
```

```python
#construction
df = pd.read_csv("newconstructionmeansalesCondo.csv")
df_constructMeanCon = cleanData(df)
df_constructMeanCon = filterYear(df_constructMeanCon)
df_constructMeanCon = combineYear(df_constructMeanCon)
df_constructMeanCon = df_constructMeanCon.rename(columns={"rename":"newConstructionMean(Condo)"})
df= pd.read_csv('newconstructionmeansaleSFR.csv')
df_constructionMeanSFR = cleanData(df)
df_constructionMeanSFR = filterYear(df_constructionMeanSFR)
df_constructionMeanSFR = combineYear(df_constructionMeanSFR)
df_constructionMeanSFR = df_constructionMeanSFR.rename(columns = {"rename":"newConstructionMean(SFR)"})
df = pd.read_csv("newconstructionsalecountcondo.csv")
df_constructionSalecondo = cleanData(df)
df_constructionSalecondo = filterYear(df_constructionSalecondo)
df_constructionSalecondo = combineYear(df_constructionSalecondo)
df_constructionSalecondo = df_constructionSalecondo.rename(columns = {"rename":"newConstructionSaleCount(Condo)"})
df = pd.read_csv("newconstructionsalecountSFR.csv")
df_constructionSaleSFR = cleanData(df)
df_constructionSaleSFR = filterYear(df_constructionSaleSFR)
df_constructionSaleSFR = combineYear(df_constructionSaleSFR)
df_constructionSaleSFR = df_constructionSaleSFR.rename(columns = {"rename":"newConstructionSaleCount(SFR)"})
df_meanCon = pd.merge(df_constructionMeanSFR,df_constructMeanCon,on =['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'Year'])
df_saleCount2 = pd.merge(df_constructionSalecondo,df_constructionSaleSFR,on =['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'Year'])

df_construct = pd.merge(df_meanCon,df_saleCount2,on =['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'Year'])
```

```python
df_sale_price = pd.merge(df_construct,df_price,on =['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'Year'])
df_sale_price
```
